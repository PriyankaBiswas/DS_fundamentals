```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomial_var = np.random.binomial(50, .8, 100)
plt.hist(binomial_var)
plt.axvline(binomial_var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial_var.mean()+binomial_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomial_var.mean()-binomial_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
chi_var = np.random.chisquare(50, 100)
plt.hist(chi_var)
plt.axvline(chi_var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chi_var.mean()+chi_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(chi_var.mean()-chi_var.std(), color='r', linestyle='dashed', linewidth=2)
```

```python
standard_normal_var = np.random.standard_normal(100)
plt.hist(standard_normal_var)
plt.axvline(standard_normal_var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standard_normal_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
standard_tvar = np.random.standard_t(50, 100)
plt.hist(standard_tvar)
plt.axvline(standard_tvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(standard_tvar.mean()+standard_tvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(standard_tvar.mean()-standard_tvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
std_exp_var = np.random.standard_exponential(100)
plt.hist(std_exp_var)
plt.axvline(std_exp_var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(std_exp_var.mean()+std_exp_var.mean(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(std_exp_var.mean()-std_exp_var.mean(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
log_var = np.random.logseries(.8, 100)
plt.hist(log_var)
plt.axvline(log_var.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(log_var.mean()+log_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(log_var.mean()-log_var.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)

plt.hist(rand1)
plt.axhline(rand1.mean(), color='r')

plt.hist(rand2)
plt.axhline(rand2.mean(), color='b')
plt.show()
```

```python
rand3 = rand1+rand2
```

```python
plt.hist(rand3)
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean()+rand3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean()-rand3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
